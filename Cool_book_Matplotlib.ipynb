```python
heart_disease=pd.read_csv("heart-disease.csv")
```

```python
heart_disease.head()
heart_disease.size
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.size
over_50
```

```python

plt.style.use('seaborn-whitegrid')


# subplots of chol, age, thalach
figgg,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

# add data to ax0:
scatter= ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter") #this changes the color scheme

# customize ax0
ax0.set(title="Heart Disease and cholestrol levels", ylabel="cholestrol")
ax0.set_xlim([50,80]) #change the x axis limits
#Add a legend to ax0:
ax0.legend(title="Target")
#ax0.legend(*scatter.legend_elements(),
                   # loc="lower left", title="Target")


#add a meanline:
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");


# add data to ax1:
scatter= ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter")

# customize ax1:
ax1.set(title="Heart Disease and Max heart rate",xlabel="Age", ylabel="Max Heart Rate")
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])                     
#Add a legend to ax1:
#ax1.legend(*scatter.legend_elements(),title="Target")

#add a meanline:
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

# Add a title to the figure
figgg.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
